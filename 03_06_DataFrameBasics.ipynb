1. Create a DataFrame whose index is the first 10 letters of the alphabet and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.

```python
import pandas as pd
import numpy as np

def listPrime(limit):
    primeList = []
    x = 2
    isPrime = True
    while len(primeList) < limit:
        sqroot = int(x ** 0.5)
        for i in range(2, sqroot + 1):
            if x % i == 0:
                isPrime = False
                break
            else:
                isPrime = True
                continue
        if isPrime:
            primeList.append(x)
        x += 1
    return primeList

listprime = listPrime(10)
primeArray = np.array(listprime)


def fibList(limit):
    fiblist = [0, 1]
    for i in range(2, limit):
        fiblist.append(fiblist[i-2] + fiblist[i-1])
    return fiblist

fiblist = fibList(10)
fibArray = np.array(fiblist)


array2 = [primeArray, fibArray]

df = pd.DataFrame(array2).T
indexList = ['a','b','c','d','e','f','g','h','i','j']
df.index = indexList
df.columns = ['primes', 'fibonacci']
df
```
